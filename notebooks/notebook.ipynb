## Beyond Risk Factors: Socioeconomic Drivers of the Global Longevity-Health Gap, 1990-2023

While countries have achieved longer lifespans, many have not extended healthspans proportionally. The project investigates how socioeconomic inequality, education and healthcare access explain differences in the gap between Life Expectancy (LE) and Health-Adjusted Life Expectancy(HALE) across 204 countries from 1990 to 2023.

This simple study aims to answer the question, **"Does socio-economic factors modify the LE-HALE gap?"**. 

To gain insights or answer the big question, here are  ___ research to guide our analysis

**Research Question**
1. How does the LE-HALE gap vary across countries over the years?
2. What is the relationship between socio-economic factors like GDP per capita, education index, health expenditure and LE-HALE gap?
3. Can we identify clusters(archetype) of countries based on their socio-economic and health characteristics?
4. Do socioeconomic variables moderate or mediate the LE-HALE relationship? 

Based on these research questions, we hypothesize that
- the LE-HALE gap is positively associated with income inequality(Gini) and negatively associated with education and health.
- socioeconomic inequality moderates the effect of health risk factors - the same level of risk exposure leads to a larger health gap in more unqual societies.
- countries cluster into distinct health-gap archetypes like
    1. Wealthy & Healthy - high LE, high HALE, low inequality

    2. Wealthy but Ailing - high LE, low HALE, high inequality

    3. Developing and Vulnerable - low LE and HALE

    4. Transitional - rapidly improving HALE
